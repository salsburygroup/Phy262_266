```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt

#set parameters and initial conditions
A = 1  # Amplitude
f = 0  # Phase
w = 1  # Frequency

# Evaluate cosine as a function of time
t = np.linspace(0, 10, 1000)  # Time  from 0 to 10 with 5 points
y = A * np.cos(w * t + f)  # Sine wave equation

# Plot my answer
plt.plot(t, y)
plt.title("Harmonic Oscillator Displacement vs time")
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
```

```python

# Plot the same sine wave for different numbers of points
num_points_list = [10, 50, 100, 200, 500, 1000]  # List of different numbers of points
for num_points in num_points_list:
    t_sampled = np.linspace(0, 10, num_points)  # Time sampled at different points
    y_sampled = A * np.cos(w * t_sampled + f)  # Sine wave equation
    plt.plot(t_sampled, y_sampled, label=f'{num_points} points')

# Customize the plot
plt.title("Harmonic Oscillator Displacement vs Time")
plt.xlabel("Time (t)")
plt.ylabel("Amplitude")
plt.legend()  # Add legend to identify each curve
plt.grid(True)  # Optional: Add grid for better readability
plt.show()
```
